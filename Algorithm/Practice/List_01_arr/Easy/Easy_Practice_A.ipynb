```python
####################################################
# [문제] 최댓값 찾기 (maxnum)
# N개의 정수가 주어진다.
# 이 중 가장 큰 값을 출력하시오.
#
# [유형] 완전 검색 (선형 탐색)
#
# [입력 예시]
# 1
# 5
# 3 17 2 9 10
#
# [출력 예시]
# #1 17
####################################################

T = int(input())
for tc in range(1, T + 1):
    N = int(input())
    arr = list(map(int, input().split()))

    max_value = arr[0]

    for i in range(N):
        if max_value < arr[i]:
            max_value = arr[i]
    
    print(f"#{tc} {max_value}")
```

```python
####################################################
# [문제] 최댓값과 인덱스 (maxnum_last_index)
# N개의 정수가 주어진다.
# 가장 큰 값과
# 그 값의 인덱스를 출력하시오.
# (인덱스는 0부터 시작)
#
# [유형] 완전 검색 + 상태 추적
#
# [입력 예시]
# 1
# 5
# 3 17 2 17 10
#
# [출력 예시]
# #1 17 1
####################################################

T = int(input())
for tc in range(1, T + 1):
    N = int(input())
    arr = list(map(int, input().split()))

    max_value = arr[0]
    max_idx = 0

    for i in range(1, N): # 이미 초기 값으로 첫 번째 인덱스 사용
        if max_value < arr[i]:
           max_value = arr[i]
           max_idx = i

    # 같은 값이 있으면 뒤에 있는 값의 인덱스
    # for i in range(1, N):
    #     if max_value <= arr[i]:
    #        max_value = arr[i]
    #        max_idx = i
    

    print(f"#{tc} {max_value} {max_idx}")
```

```python
####################################################
# [문제] 최댓값과 최솟값 차이 (min_max)
# N개의 정수가 주어진다.
# 가장 큰 값과
# 가장 작은 값의 차이를 출력하시오.
#
# [유형] 완전 검색 (다중 상태)
#
# [입력 예시]
# 1
# 5
# 7 4 2 9 1
#
# [출력 예시]
# #1 8
####################################################

T = int(input())
for tc in range(1, T + 1):
    N = int(input())
    arr = list(map(int, input().split()))

    max_value = arr[0]
    min_value = arr[0]

    for i in range(1, N):
        if max_value < arr[i]:
            max_value = arr[i]

        if min_value > arr[i]:
            min_value = arr[i]

    print(f"#{tc} {max_value - min_value}")
```

```python
####################################################
# [문제] 고정 구간 비교 (view)
# N개의 건물이 일렬로 서 있다.
# i번째 건물의 높이가
# 양쪽 두 칸의 건물보다 모두 높으면
# 조망권이 확보되었다고 한다.
#
# 조망권이 확보된 건물의 개수를 출력하시오.
#
# [유형] 완전 검색 (고정 범위 비교)
#
# [입력 예시]
# 1
# 8
# 0 0 3 5 2 4 0 0
#
# [출력 예시]
# #1 1
####################################################

T = int(input())
for tc in range(1, T + 1):
    N = int(input())
    h = list(map(int, input().split()))

    count = 0

    # 조망권의 요구 조건 양 옆 두 칸 / 조건마다 달라짐
    for i in range(2, N - 2): 
        
        # 첫 번째 값을 기준값으로 둔다
        max_near = h[i - 2]

        if max_near < h[i - 1]:
            max_near = h[i - 1]

        if max_near < h[i + 1]:
            max_near = h[i + 1]

        if max_near < h[i + 2]:
            max_near = h[i + 2]

        if h[i] > max_near:
            count += 1

    print(f"#{tc} {count}")
```

```python
####################################################
# [문제] 낙차 계산 (gravity)
# N개의 상자가 일렬로 놓여 있다.
# 각 위치에서 오른쪽에 있는 상자 중
# 자신보다 낮은 상자의 개수를 센다.
#
# 그 중 가장 큰 값을 출력하시오.
#
# [유형] 완전 검색 (이중 반복)
#
# [입력 예시]
# 1
# 9
# 7 4 2 0 0 6 0 7 0
#
# [출력 예시]
# #1 7
####################################################

T = int(input())
for tc in range(1, T + 1):
    N = int(input())
    boxes = list(map(int, input().split()))

    max_drop = 0
    
    # 마지막 상자는 오른쪽에 비교할 상자가 없으므로 제외
    for i in range(N-1):

        cnt = 0 # 자신보다 낮은 상자 갯수 카운트

        for j in range(i + 1, N): # 각 위치에서 오른쪽으로 N까지

            if boxes[i] > boxes[j]:
               cnt += 1

        # 지금까지 센 카운트가 최댓값 보다 크면 교체
        if max_drop < cnt:
            max_drop = cnt

    print(f"#{tc} {max_drop}")
```
