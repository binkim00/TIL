```python
####################################################
# [문제] 탐욕 알고리즘 (Greedy)
# 배열이 주어진다.
# 현재 배열에서
# 가장 큰 값을 찾아 출력하시오.
#
# [유형] 탐욕 (현재 최선 선택)
#
# [입력 예시]
# 1
# 5
# 3 7 2 9 1
#
# [출력 예시]
# #1 9
####################################################

T = int(input())
for tc in range(1, T + 1):
    N = int(input())
    arr = list(map(int, input().split()))

    max_value = arr[0]

    for i in range(1, N):
        if max_value < arr[i]:
            max_value = arr[i]

    print(f"#{tc} {max_value}")
```

```python
####################################################
# [문제] 탐욕 + 시뮬레이션
#
# [요구 조건 정리]
# 1️⃣ 배열에서 "가장 큰 값을 가진 원소"를 찾는다
#    - 값만이 아니라 '위치(인덱스)'가 필요함
#
# 2️⃣ 배열에서 "가장 작은 값을 가진 원소"를 찾는다
#    - 마찬가지로 위치(인덱스)를 알아야 함
#
# 3️⃣ 위에서 찾은 두 원소에 대해
#    - 가장 큰 값  → 1 감소
#    - 가장 작은 값 → 1 증가
#    → 배열의 상태를 실제로 변경해야 함 (시뮬레이션)
#
# 4️⃣ 변경된 배열을 기준으로
#    - 최고점(최댓값)
#    - 최저점(최솟값)
#    을 다시 구한다
#
# 5️⃣ 최종 출력
#    - (최고점 - 최저점)
#
# ⚠️ 주의
# - 단순히 (최댓값 - 1) - (최솟값 + 1) 로 계산하면
#   이 문제에서는 우연히 맞을 수 있으나,
#   탐욕 + 시뮬레이션 문제의 올바른 풀이 방식은 아님
# [입력 예시]
# 1
# 5
# 7 4 2 9 1
#
# [출력 예시]
# #1 6
####################################################

T = int(input())
for tc in range(1, T + 1):
    N = int(input())
    arr = list(map(int, input().split()))

    # 1️⃣ 최댓값 / 최솟값 위치 찾기
    max_idx = 0
    min_idx = 0

    for i in range(1, N):
        if arr[max_idx] < arr[i]:
            max_idx = i
        if arr[min_idx] > arr[i]:
            min_idx = i

    # 2️⃣ 실제로 값 변경 (시뮬레이션)
    arr[max_idx] -= 1
    arr[min_idx] += 1

    # 3️⃣ 변경된 배열 기준으로 다시 최댓값/최솟값
    max_num = arr[0]
    min_num = arr[0]

    for i in range(1, N):
        if max_num < arr[i]:
            max_num = arr[i]
        if min_num > arr[i]:
            min_num = arr[i]

    print(f"#{tc} {max_num - min_num}")
```

```python
####################################################
# [문제] 탐욕 + 반복 시뮬레이션
# 배열에서
# 가장 큰 값에서 1을 빼고
# 가장 작은 값에 1을 더하는 작업을
# K번 수행한다.
# K번 수행 후,
# 최고점과 최저점의 차이를 출력하시오.
#
# [유형] 탐욕 + 시뮬레이션
#
# [입력 예시]
# 1
# 5 2
# 7 4 2 9 1
#
# [출력 예시]
# #1 5
####################################################

T = int(input())
for tc in range(1, T + 1):
    N, K = map(int, input().split())
    arr = list(map(int, input().split()))

    # K번 = "작업 1회"를 K번 반복
    for _ in range(K):

        # ⭐ 매 회차마다 다시 찾기
        # (이전 회차에서의 max/min 위치는 더 이상 의미 없음)
        max_idx = 0
        min_idx = 0

        # 현재 배열 상태 기준으로
        # 최댓값 / 최솟값 위치 찾기
        for i in range(N):
            if arr[max_idx] < arr[i]:
                max_idx = i
            if arr[min_idx] > arr[i]:
                min_idx = i

        # ⭐ 탐색이 끝난 뒤에 딱 1번만 값 이동
        # (이게 "시뮬레이션 1회")
        arr[max_idx] -= 1
        arr[min_idx] += 1

    # 모든 이동이 끝난 최종 배열에서
    # 최고점 / 최저점 다시 계산
    max_num = arr[0]
    min_num = arr[0]

    for i in range(N):
        if max_num < arr[i]:
            max_num = arr[i]
        if min_num > arr[i]:
            min_num = arr[i]

    print(f"#{tc} {max_num - min_num}")
```

```python
####################################################
# [문제] 카운트 배열 (Counting) 개수 세기
# 0부터 9까지의 숫자로 이루어진 배열이 주어진다.
# 각 숫자가 몇 번씩 등장했는지 출력하시오.
#
# [유형] 카운트 배열
#
# [입력 예시]
# 1
# 7
# 1222333
#
# [출력 예시]
# #1 [0, 1, 3, 3, 0, 0, 0, 0, 0, 0]
####################################################
T = int(input())

for tc in range(1, T + 1):
    N = int(input())
    arr = list(map(int, input().strip()))   # 문자열 '1222333' -> [1,2,2,2,3,3,3]
    # ⚠️ N은 길이 확인용(여기서는 굳이 안 써도 됨)

    cnt = [0] * 10  # 0~9까지 각각의 횟수를 저장할 칸 10개 만들기

    # ⭐ 여기 한 번이면 "개수 세기" 끝
    for x in arr:
        # x 자체가 숫자(0~9)라서, 그 숫자 위치(cnt[x])에 +1 하면 됨
        cnt[x] += 1

    print(f"#{tc} {cnt}")
```

```python
####################################################
# [문제] 카운트 배열 (Counting) 최빈값
# 0~9 숫자로 이루어진 배열이 주어진다.
# 가장 많이 등장한 숫자와 등장 횟수를 출력하시오.
# (개수가 같다면 숫자가 큰 쪽)
#
# [풀이 핵심]
# 1) cnt[숫자]에 등장 횟수 누적
# 2) cnt를 0~9까지 확인하면서
#    - 더 큰 횟수면 갱신
#    - 횟수가 같으면 숫자가 더 큰 쪽으로 갱신
####################################################

T = int(input())
for tc in range(1, T + 1):
    N = int(input())
    arr = list(map(int, input().strip()))

    cnt = [0] * 10
    max_num = 0
    max_cnt = 0

    # 1) 개수 세기
    for x in arr:
        # x 자체가 숫자(0~9)라서, 그 숫자 위치(cnt[x])에 +1 하면 됨        
        cnt[x] += 1

    # 0부터 9까지 모든 숫자를 하나씩 확인
    for i in range(10):

        # 현재 숫자(i)가 지금까지의 최댓횟수보다 더 많이 나왔다면
        if max_cnt < cnt[i]:
            max_cnt = cnt[i]    # 최댓 등장 횟수 갱신
            max_num = i         # 해당 숫자로 최빈값 갱신

        # 등장 횟수가 같다면
        # (문제 조건: 개수가 같을 때는 숫자가 큰 쪽 선택)
        elif max_cnt == cnt[i] and max_num < i:
            max_num = i         # 더 큰 숫자로 최빈값 갱신


    print(f"#{tc} {max_num} {max_cnt}")
```

```python
####################################################
# [문제 설명]
# 숫자들이 들어있는 배열이 있다.
# 이 배열에서 서로 다른 숫자 두 개를 골라서
# 두 숫자의 차이의 절댓값 중
# 가장 큰 값을 구하시오.
#
# 예)
# [1, 3, 5, 9]
# 가능한 경우:
# |1-3|, |1-5|, |1-9|,
# |3-5|, |3-9|,
# |5-9|
# 이 중 가장 큰 값은 8
####################################################

T = int(input())
for tc in range(1, T + 1):
    N = int(input())                    # 숫자 개수
    arr = list(map(int, input().split()))  # 숫자 배열

    max_diff = 0                        # 가장 큰 차이를 저장할 변수

    # 첫 번째 숫자 고르기
    for i in range(N):
        # 두 번째 숫자 고르기 (첫 번째 다음부터)
        for j in range(i + 1, N):

            # 두 숫자의 차이 구하기
            diff = arr[i] - arr[j]

            # 차이가 음수면 양수로 바꾸기 (절댓값 만들기)
            if diff < 0:
                diff = -diff

            # 지금까지의 최대값보다 크면 바꾸기
            if max_diff < diff:
                max_diff = diff

    print(f"#{tc} {max_diff}")
```

```python
####################################################
# [문제] 부분 수열 (투 포인터) / 문자열 부분 수열
# 문자열 A와 문자열 B가 주어진다.
# B가 A의 부분 수열이면 1,
# 아니면 0을 출력하시오.
#
# [유형] 부분 수열
#
# [입력 예시]
# 1
# abcdef
# ace
#
# [출력 예시]
# #1 1
####################################################
T = int(input())
for tc in range(1, T + 1):
    A = input().strip()
    B = input().strip()

    idx = 0     # B에서 현재 맞춰야 할 위치
    result = 0

    for ch in A:
        # B[idx]가 존재하면 비교
        if ch == B[idx]:
            idx += 1

            # 다음 문자가 없으면(B 끝까지 맞췄으면)
            if B[idx:idx+1] == "":
                result = 1
                break

    print(f"#{tc} {result}")

```
