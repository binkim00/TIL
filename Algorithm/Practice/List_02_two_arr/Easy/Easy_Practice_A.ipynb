```python
####################################################
# [문제 1] 2차원 배열 순회
# N x N 크기의 2차원 배열이 주어진다.
#
# 배열의 모든 원소를 순회하며
# 값의 총합을 출력하시오.
#
# [입력 예시]
# 1
# 3
# 1 2 3
# 4 5 6
# 7 8 9
#
# [출력 예시]
# #1 45
####################################################

T = int(input())
for tc in range(1, T + 1):
    N = int(input())
    arr = [list(map(int, input().split())) for _ in range(N)]

    total = 0

    for i in range(N):
        for j in range(N):
            total += arr[i][j]

    print(f"#{tc} {total}")
```

```python
####################################################
# [문제 2] 행 합 최댓값
# N x N 크기의 2차원 배열이 주어진다.
#
# i ↓  (행, row, 가로줄)
# j →  (열, column, 세로칸)
#
# 각 행의 합을 구한 뒤,
# 그 중 최댓값을 출력하시오.
#
# [입력 예시]
# 1
# 3
# 1 2 3
# 3 3 3
# 0 10 0
#
# [출력 예시]
# #1 10
####################################################

T = int(input())
for tc in range(1, T + 1):
    N = int(input())
    arr = [list(map(int, input().split())) for _ in range(N)]

    max_sum = 0

    for i in range(N):

        row_sum = 0
    
        for j in range(N):
            row_sum += arr[i][j]
    
        if max_sum < row_sum:
            max_sum = row_sum

    print(f"#{tc} {max_sum}")
```

```python
####################################################
# [문제 3] 열 합 최댓값
# N x N 크기의 2차원 배열이 주어진다.
#
# 각 열의 합을 구한 뒤,
# 그 중 최댓값을 출력하시오.
#
# [입력 예시]
# 1
# 3
# 1 2 3
# 3 3 3
# 0 10 0
#
# [출력 예시]
# #1 15
####################################################

T = int(input())
for tc in range(1, T + 1):
    N = int(input())
    arr = [list(map(int, input().split())) for _ in range(N)]

    max_sum = 0

    for j in range(N):

        col_sum = 0

        for i in range(N):
            col_sum += arr[i][j]
        
        if max_sum < col_sum:
            max_sum = col_sum

    print(f"#{tc} {max_sum}")
```

```python
####################################################
# [문제 4] 델타 탐색 (상하좌우)
# N x M 크기의 2차원 배열이 주어진다.
#
# 각 칸을 기준으로
# 자기 자신 + 상하좌우 1칸의 합을 구한다.
#
# 이 중 최댓값을 출력하시오.
#
# [입력 예시]
# 1
# 3 3
# 1 2 3
# 4 5 6
# 7 8 9
#
# [출력 예시]
# #1 29
####################################################
T = int(input())

for tc in range(1, T + 1):
    N, M = map(int, input().split())
    arr = [list(map(int, input().split())) for _ in range(N)]

    # 위, 아래, 왼, 오 이동
    di = [-1, 1, 0, 0]
    dj = [0, 0, -1, 1]

    best = 0  # 가장 큰 합

    for i in range(N):          # 모든 행
        for j in range(M):      # 모든 열

            total = arr[i][j]   # ⭐ 내 자리 값부터

            for d in range(4):  # 네 방향 확인
                ni = i + di[d]  # 이동한 행
                nj = j + dj[d]  # 이동한 열

                # 배열 안에 있으면
                if 0 <= ni < N and 0 <= nj < M:
                    total += arr[ni][nj]

            # 가장 큰 값이면 저장
            if best < total:
                best = total

    print(f"#{tc} {best}")

    # 방향에 거리 추가
    # for d in range(4):          # 네 방향
    # for k in range(1, K+1): # 1칸 ~ K칸
    #     ni = i + di[d] * k
    #     nj = j + dj[d] * k

    #     if 0 <= ni < N and 0 <= nj < M:
    #         total += arr[ni][nj]


```

```python
####################################################
# [문제 5] 델타 탐색 (대각선)
# N x N 크기의 2차원 배열이 주어진다.
#
# 각 칸을 기준으로
# 자기 자신 + 대각선 4방향의 합을 구한다.
#
# 이 중 최댓값을 출력하시오.
#
# [입력 예시]
# 1
# 3
# 1 2 3
# 4 5 6
# 7 8 9
#
# [출력 예시]
# #1 25
####################################################

T = int(input())

for tc in range(1, T + 1):
    N = int(input())
    arr = [list(map(int, input().split())) for _ in range(N)]

    # 대각선 4방향
    di = [-1, -1,  1,  1]
    dj = [-1,  1, -1,  1]


    best = 0  # 가장 큰 합

    for i in range(N):          # 모든 행
        for j in range(N):      # 모든 열

            total = arr[i][j]   # ⭐ 내 자리 값부터

            for d in range(4):  # 대각선 방향 확인
                ni = i + di[d]  # 이동한 행
                nj = j + dj[d]  # 이동한 열

                # 배열 안에 있으면
                if 0 <= ni < N and 0 <= nj < N:
                    total += arr[ni][nj]

            # 가장 큰 값이면 저장
            if best < total:
                best = total

    print(f"#{tc} {best}")
```
