```python
####################################################
# [문제 2] 행 합 최댓값
# N x N 크기의 2차원 배열이 주어진다.
#
# i ↓  (행, row, 가로줄)
# j →  (열, column, 세로칸)
#
# 각 행의 합을 구한 뒤,
# 그 중 최댓값을 출력하시오.
#
# [입력 예시]
# 1
# 3
# 1 2 3
# 3 3 3
# 0 10 0
#
# [출력 예시]
# #1 10
####################################################

T = int(input())
for tc in range(1, T + 1):
    N = int(input())
    arr = [list(map(int, input().split())) for _ in range(N)]

    max_sum = 0

    for i in range(N): # 행 하나 선택

        row_sum = 0
    
        for j in range(N): # 그 행의 모든 열
            row_sum += arr[i][j]
        # 여기까지가 "한 행"
    
        if max_sum < row_sum:
            max_sum = row_sum

    print(f"#{tc} {max_sum}")
```

```python
####################################################
# [문제] 델타 탐색 (8방향)
# N x N 크기의 2차원 배열이 주어진다.
#
# 각 칸을 기준으로
# 자기 자신 + 주변 8방향(상하좌우 + 대각선 4개) 1칸의 합을 구한다.
# 이 합들 중 최댓값을 출력하시오.
#
# [유형] 델타 탐색 (8방향)
#
# [입력 예시]
# 1
# 3
# 1 2 3
# 4 5 6
# 7 8 9
#
# [출력 예시]
# #1 45
####################################################

T = int(input())
for tc in range(1, T + 1):
    N = int(input())
    arr = [list(map(int, input().split())) for _ in range(N)]

    # 8방향(상, 우상, 우, 우하, 하, 좌하, 좌, 좌상)
    di = [-1, -1,  0,  1, 1,  1,  0, -1]
    dj = [ 0,  1,  1,  1, 0, -1, -1, -1]

    best = 0

    for i in range(N):
        for j in range(N):
            total = arr[i][j]  # 자기 자신

            # 주변 8방향 1칸 더하기
            for d in range(8):
                ni = i + di[d]
                nj = j + dj[d]

                if 0 <= ni < N and 0 <= nj < N:
                    total += arr[ni][nj]

            if best < total:
                best = total

    print(f"#{tc} {best}")


####################################################
# [거리(몇 칸) 추가하는 법]
#
# 만약 문제에서 "8방향으로 K칸까지 더하라" 라고 나오면,
# 아래처럼 방향(d) 안에서 거리(k)를 한 번 더 돌리면 된다.
#
# for d in range(8):              # 방향 8개
#     for k in range(1, K + 1):   # 1칸 ~ K칸
#         ni = i + di[d] * k
#         nj = j + dj[d] * k
#         if 0 <= ni < N and 0 <= nj < N:
#             total += arr[ni][nj]
####################################################
```
