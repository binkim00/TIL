```python
####################################################
# [문제] 고정 구간 비교 (view)
# N개의 건물이 일렬로 서 있다.
# i번째 건물의 높이가
# 양쪽 두 칸의 건물보다 모두 높으면
# 조망권이 확보되었다고 한다.
#
# 조망권이 확보된 건물의 개수를 출력하시오.
#
# [유형] 완전 검색 (고정 범위 비교)
#
# [입력 예시]
# 1
# 8
# 0 0 3 5 2 4 0 0
#
# [출력 예시]
# #1 1
####################################################

T = int(input())
for tc in range(1, T + 1):
    N = int(input())
    h = list(map(int, input().split()))

    count = 0

    # 조망권의 요구 조건 양 옆 두 칸 / 조건마다 달라짐
    for i in range(2, N - 2): 
        
        # 첫 번째 값을 기준값으로 둔다
        max_near = h[i - 2]

        if max_near < h[i - 1]:
            max_near = h[i - 1]

        if max_near < h[i + 1]:
            max_near = h[i + 1]

        if max_near < h[i + 2]:
            max_near = h[i + 2]

        if h[i] > max_near:
            count += 1

    print(f"#{tc} {count}")
```

```python
####################################################
# [문제] 낙차 계산 (gravity)
# N개의 상자가 일렬로 놓여 있다.
# 각 위치에서 오른쪽에 있는 상자 중
# 자신보다 낮은 상자의 개수를 센다.
#
# 그 중 가장 큰 값을 출력하시오.
#
# [유형] 완전 검색 (이중 반복)
#
# [입력 예시]
# 1
# 9
# 7 4 2 0 0 6 0 7 0
#
# [출력 예시]
# #1 7
####################################################

T = int(input())
for tc in range(1, T + 1):
    N = int(input())
    boxes = list(map(int, input().split()))

    max_drop = 0
    
    # 마지막 상자는 오른쪽에 비교할 상자가 없으므로 제외
    for i in range(N-1):

        cnt = 0 # 자신보다 낮은 상자 갯수 카운트

        for j in range(i + 1, N): # 각 위치에서 오른쪽으로 N까지

            if boxes[i] > boxes[j]:
               cnt += 1

        # 지금까지 센 카운트가 최댓값 보다 크면 교체
        if max_drop < cnt:
            max_drop = cnt

    print(f"#{tc} {max_drop}")
```

```python
####################################################
# [문제] 탐욕 + 반복 시뮬레이션
# 배열에서
# 가장 큰 값에서 1을 빼고
# 가장 작은 값에 1을 더하는 작업을
# K번 수행한다.
# K번 수행 후,
# 최고점과 최저점의 차이를 출력하시오.
# [요구 조건 정리]
# 1️⃣ 배열에서 "가장 큰 값을 가진 원소"를 찾는다
#    - 값만이 아니라 '위치(인덱스)'가 필요함
#
# 2️⃣ 배열에서 "가장 작은 값을 가진 원소"를 찾는다
#    - 마찬가지로 위치(인덱스)를 알아야 함
#
# 3️⃣ 위에서 찾은 두 원소에 대해
#    - 가장 큰 값  → 1 감소
#    - 가장 작은 값 → 1 증가
#    → 배열의 상태를 실제로 변경해야 함 (시뮬레이션)
#
# 4️⃣ 변경된 배열을 기준으로
#    - 최고점(최댓값)
#    - 최저점(최솟값)
#    을 다시 구한다
#
# 5️⃣ 최종 출력
#    - (최고점 - 최저점)
#
# [유형] 탐욕 + 시뮬레이션
#
# [입력 예시]
# 1
# 5 2
# 7 4 2 9 1
#
# [출력 예시]
# #1 5
####################################################

T = int(input())
for tc in range(1, T + 1):
    N, K = map(int, input().split())
    arr = list(map(int, input().split()))

    # K번 = "작업 1회"를 K번 반복
    for _ in range(K):

        # ⭐ 매 회차마다 다시 찾기
        # (이전 회차에서의 max/min 위치는 더 이상 의미 없음)
        max_idx = 0
        min_idx = 0

        # 현재 배열 상태 기준으로
        # 최댓값 / 최솟값 위치 찾기
        for i in range(N):
            if arr[max_idx] < arr[i]:
                max_idx = i
            if arr[min_idx] > arr[i]:
                min_idx = i

        # ⭐ 탐색이 끝난 뒤에 딱 1번만 값 이동
        # (이게 "시뮬레이션 1회")
        arr[max_idx] -= 1
        arr[min_idx] += 1

    # 모든 이동이 끝난 최종 배열에서
    # 최고점 / 최저점 다시 계산
    max_num = arr[0]
    min_num = arr[0]

    for i in range(N):
        if max_num < arr[i]:
            max_num = arr[i]
        if min_num > arr[i]:
            min_num = arr[i]

    print(f"#{tc} {max_num - min_num}")
```

```python
####################################################
# [문제] 카운트 배열 (Counting) 최빈값
# 0~9 숫자로 이루어진 배열이 주어진다.
# 가장 많이 등장한 숫자와 등장 횟수를 출력하시오.
# (개수가 같다면 숫자가 큰 쪽)
#
# [풀이 핵심]
# 1) cnt[숫자]에 등장 횟수 누적
# 2) cnt를 0~9까지 확인하면서
#    - 더 큰 횟수면 갱신
#    - 횟수가 같으면 숫자가 더 큰 쪽으로 갱신
####################################################

T = int(input())
for tc in range(1, T + 1):
    N = int(input())
    arr = list(map(int, input().strip()))

    cnt = [0] * 10
    max_num = 0
    max_cnt = 0

    # 1) 개수 세기
    for x in arr:
        # x 자체가 숫자(0~9)라서, 그 숫자 위치(cnt[x])에 +1 하면 됨        
        cnt[x] += 1

    # 0부터 9까지 모든 숫자를 하나씩 확인
    for i in range(10):

        # 현재 숫자(i)가 지금까지의 최댓횟수보다 더 많이 나왔다면
        if max_cnt < cnt[i]:
            max_cnt = cnt[i]    # 최댓 등장 횟수 갱신
            max_num = i         # 해당 숫자로 최빈값 갱신

        # 등장 횟수가 같다면
        # (문제 조건: 개수가 같을 때는 숫자가 큰 쪽 선택)
        elif max_cnt == cnt[i] and max_num < i:
            max_num = i         # 더 큰 숫자로 최빈값 갱신


    print(f"#{tc} {max_num} {max_cnt}")
```
