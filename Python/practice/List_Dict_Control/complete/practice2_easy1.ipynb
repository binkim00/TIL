```python
############## 주의 ##############
# 문자열 길이 합 구하기
# input 함수는 사용하지 않습니다.
# 여기에 코드를 작성하여 함수를 완성합니다.
def sum_length(words):
    total = 0              # 문자열 길이를 누적해서 더할 변수 (초기값 0)
    for c in words:        # words 리스트에서 문자열 하나씩 꺼내기
        total += len(c)    # 현재 문자열 c의 길이를 total에 더함
    return total           # 모든 문자열 길이를 더한 최종 결과 반환


############## 테스트 코드 삭제 금지 #################
print(sum_length(["hi", "hello", "a"]))   # 8
print(sum_length(["aa", "bb"]))           # 4
#####################################################
```

```python
############## 주의 ##############
# 대문자로만 구성된 문자열 개수 세기
# input 함수는 사용하지 않습니다.
# 여기에 코드를 작성하여 함수를 완성합니다.
def count_upper(words):
    total = 0              # 대문자로만 된 문자열 개수를 셀 변수
    for c in words:        # words 리스트에서 문자열 하나씩 꺼내기
        if c.isupper():    # 문자열 c가 전부 대문자인지 확인
            total += 1     # 조건을 만족하면 개수 1 증가
    return total           # 대문자 문자열 총 개수 반환


############## 테스트 코드 삭제 금지 #################
print(count_upper(["ABC", "Abc", "DEF"]))  # 2
print(count_upper(["hello", "WORLD"]))     # 1
#####################################################
```

```python
############## 주의 ##############
# 숫자 리스트에서 문자열로 변환한 값 리스트 만들기
# input 함수는 사용하지 않습니다.
# 여기에 코드를 작성하여 함수를 완성합니다.
def number_to_string(numbers):
    numlist = []                 # 문자열로 변환된 값을 담을 빈 리스트
    for num in numbers:          # numbers 리스트에서 숫자 하나씩 꺼내기
        numlist.append(str(num)) # 숫자를 문자열로 변환하여 리스트에 추가
    return numlist               # 변환된 문자열 리스트 반환


############## 테스트 코드 삭제 금지 #################
print(number_to_string([1, 2, 3]))   # ['1', '2', '3']
print(number_to_string([10, 20]))    # ['10', '20']
#####################################################
```

```python
############## 주의 ##############
# 한번에 정답!
# 특정 글자를 포함한 단어만 리스트로 만들기
# input 함수는 사용하지 않습니다.
def filter_word(words, letter):
    # 여기에 코드를 작성하여 함수를 완성합니다.
    word_list = []
    for word in words:
        if letter in word:
            word_list.append(word)
    return word_list


############## 테스트 코드 삭제 금지 #################
print(filter_word(["apple", "banana", "cherry"], "a"))
# ['apple', 'banana']
#####################################################
```

```python
############## 주의 ##############
# 점수가 80점 이상인 학생 이름만 리스트로 반환
# input 함수는 사용하지 않습니다.
def high_score_students(scores):
    # 여기에 코드를 작성하여 함수를 완성합니다.
    pass_students = []                  # 조건을 만족한 학생 이름을 담을 리스트
    for name, score in scores.items():  # 딕셔너리에서 이름과 점수를 함께 꺼냄
        if score >= 80:                 # 점수가 80 이상인지 확인
            pass_students.append(name)  # 학생 이름을 결과 리스트에 추가
    return pass_students                # 최종 결과 반환


############## 테스트 코드 삭제 금지 #################
scores = {"kim": 80, "lee": 95, "park": 70}
print(high_score_students(scores))  # ['kim', 'lee']
#####################################################
```

```python
############## 주의 ##############
# 회사별 인원 수 세기
# input 함수는 사용하지 않습니다.
def count_by_company(users):
    # 여기에 코드를 작성하여 함수를 완성합니다.    
    result = {}                            # 회사별 인원 수를 저장할 딕셔너리
    for user in users:                     # users 리스트에서 사용자 하나씩 꺼내기
        company = user["company"]          # 현재 사용자의 회사명 가져오기

        if company not in result:          # 아직 딕셔너리에 없는 회사라면
            result[company] = 1            # 처음 등장 → 1명으로 시작
        else:
            result[company] += 1           # 이미 있으면 인원 수 +1

    return result                          # 회사별 인원 수 딕셔너리 반환


############## 테스트 코드 삭제 금지 #################
users = [
    {"name": "철수", "company": "A"},
    {"name": "영희", "company": "B"},
    {"name": "민수", "company": "A"},
]
print(count_by_company(users))  # {'A': 2, 'B': 1}
#####################################################
```

```python
############## 주의 ##############
# 얼떨결에 맞음
# 문자열 리스트에서 가장 긴 문자열의 길이 구하기
# input 함수는 사용하지 않습니다.
def longest_length(words):
    # 여기에 코드를 작성하여 함수를 완성합니다.
    max_len = 0
    ch_len = 0
    
    for ch in words:
        ch_len = len(ch)
        if ch_len > max_len:
            max_len = ch_len
    return max_len


############## 테스트 코드 삭제 금지 #################
print(longest_length(["hi", "hello", "bye"]))  # 5
#####################################################
```

```python
############## 주의 ##############
# 한 번에 맞음
# 숫자 리스트에서 중복 제거한 리스트 만들기
# input 함수는 사용하지 않습니다.
def remove_duplicate(numbers):
    # 여기에 코드를 작성하여 함수를 완성합니다.
    num_list = []
    for num in numbers:
        if num not in num_list:
            num_list.append(num)

    return num_list


############## 테스트 코드 삭제 금지 #################
print(remove_duplicate([1, 2, 2, 3, 1]))  # [1, 2, 3]
#####################################################

# set 버전/ 입력 순서 유지하면 사용 x/ 정렬된 결과를 요구하고 입력 순서가 중요하지 않을 경우
############## 주의 ##############
# 숫자 리스트에서 중복 제거 + 오름차순 정렬
# input 함수는 사용하지 않습니다.
def remove_duplicate_sorted(numbers):
    return sorted(set(numbers))


############## 테스트 코드 삭제 금지 #################
print(remove_duplicate_sorted([1, 2, 2, 3, 1]))  # [1, 2, 3]
#####################################################

```

```python
############## 주의 ##############
# 한 번에 정답
# 숫자 리스트에서 홀수는 제거하고 짝수만 제곱하여 반환
# input 함수는 사용하지 않습니다.
def square_even(numbers):
    # 여기에 코드를 작성하여 함수를 완성합니다.
    even_sqare_list = []
    for num in numbers:
        if num % 2 == 0:
            even_sqare_list.append(num ** 2)

    return even_sqare_list


############## 테스트 코드 삭제 금지 #################
print(square_even([1, 2, 3, 4, 5]))  # [4, 16]
#####################################################
```
