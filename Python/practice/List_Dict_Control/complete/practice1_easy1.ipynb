```python
numbers = [3, 7, 2, 9, 4]

# 1. 첫 번째 값을 기준값(max_value)으로 저장한다
# 2. 리스트를 반복하면서 각 값을 기준값과 비교한다
# 3. 더 큰 값이 나오면 기준값을 갱신한다

# ↓ 여기에 코드 작성
def max_value(nums):
    max_num = nums[0]
    for num in numbers:
        if num > max_num:
            max_num = num
        
    return max_num

print(max_value(numbers))  # 9
```

```python
############## 주의 ##############
# 짝수 개수 세기
# input 함수는 사용하지 않습니다.
def count_even(numbers):
    count = 0
    for num in numbers:
        if num % 2 == 0:
            count += 1
    return count
    # 여기에 코드를 작성하여 함수를 완성합니다.


############## 테스트 코드 삭제 금지 #################
print(count_even([1, 2, 3, 4, 6, 7]))  # 3
print(count_even([2, 4, 6, 8]))        # 4
print(count_even([1, 3, 5]))           # 0
#####################################################
```

```python
############## 주의 ##############
# 리스트 합 구하기
# input 함수는 사용하지 않습니다.
def sum_numbers(numbers):
    addresult = 0
    for num in numbers:
        addresult += num
    return addresult
    # 여기에 코드를 작성하여 함수를 완성합니다.


############## 테스트 코드 삭제 금지 #################
print(sum_numbers([3, 5, 2, 7]))    # 17
print(sum_numbers([1, 1, 1, 1]))    # 4
#####################################################
```

```python
############## 주의 ##############
# 1회 틀림
# 평균보다 큰 값 개수
# input 함수는 사용하지 않습니다.
def count_above_avg(numbers):
    count = 0
    total = 0
    result = 0
    for num in numbers:
        count += 1
        total += num

    avg = total / count

    for num in numbers:
        if num > avg:
            result += 1
        
    return result

    # 여기에 코드를 작성하여 함수를 완성합니다.


############## 테스트 코드 삭제 금지 #################
print(count_above_avg([10, 20, 30, 40]))  # 2
print(count_above_avg([5, 5, 5, 5]))      # 0
#####################################################
```

```python
############## 주의 ##############
# 1회 틀림
# 최솟값 찾기
# input 함수는 사용하지 않습니다.
# 파이썬 내장 함수 min은 사용하지 않습니다.

"""
1) min_num = 0 으로 시작
[9, 3, 7, 1, 5]는 우연히 맞을 수 있음
하지만 [-3, -5, -1] 같은 케이스에서 0이 계속 최소 후보로 남는 이상한 상황이 생깁니다.

2) min_num = numbers (리스트 자체)
리스트와 숫자를 비교하게 돼서 타입 에러/논리 오류.

3) min_num = numbers[1] 처럼 첫 원소를 빼먹음
첫 값이 최솟값인 케이스에서 틀릴 수 있음.

- 한 줄로 외우는 규칙
최댓값/최솟값 문제에서 기준값은 “리스트의 첫 값”으로 시작한다.

"""
def find_min(numbers):
    min_num = numbers[0] # 이 이줄을 계속 틀림
    for num in numbers:
        if num < min_num:
            min_num = num

    return min_num
    # 여기에 코드를 작성하여 함수를 완성합니다.


############## 테스트 코드 삭제 금지 #################
print(find_min([9, 3, 7, 1, 5]))  # 1
print(find_min([100, 50, 70]))    # 50
#####################################################
```

```python
############## 주의 ##############
# 완성 못함
# 가장 많이 등장한 값 찾기
# input 함수는 사용하지 않습니다.
# 여기에 코드를 작성하여 함수를 완성합니다.
def find_mode(numbers):
    counts = {}          
    # 각 숫자가 몇 번 나왔는지 저장할 딕셔너리
    # 예: {2: 3} → 숫자 2가 3번 나옴

    max_count = 0        
    # 지금까지 확인한 숫자 중 가장 많이 나온 횟수

    mode = numbers[0]    
    # 가장 많이 나온 숫자를 임시로 저장
    # 일단 첫 번째 숫자를 기준으로 잡음

    # 1단계: 숫자별 등장 횟수 세기
    for num in numbers:      
        # numbers 리스트에서 숫자를 하나씩 꺼냄

        if num in counts:    
            # 이미 등장한 숫자라면
            counts[num] += 1 
            # 등장 횟수를 1 증가
        else:
            # 처음 등장한 숫자라면
            counts[num] = 1  
            # 등장 횟수를 1로 시작

    # 2단계: 가장 많이 등장한 숫자 찾기
    for num in counts:       
        # 등장한 숫자들만 하나씩 확인

        if counts[num] > max_count:
            # 현재 숫자의 등장 횟수가
            # 지금까지의 최대 횟수보다 크면

            max_count = counts[num]
            # 최대 등장 횟수 갱신

            mode = num
            # 가장 많이 등장한 숫자 갱신

    return mode
    # 최종적으로 가장 많이 등장한 숫자를 반환



############## 테스트 코드 삭제 금지 #################
print(find_mode([1, 2, 2, 3, 3, 3, 4]))  # 3
print(find_mode([5, 5, 6, 6, 5]))        # 5
#####################################################
```

```python
############## 주의 ##############
# 조건 필터 + 가공
# 짝수만 제곱해서 리스트에 담기
# input 함수는 사용하지 않습니다.
# 여기에 코드를 작성하여 함수를 완성합니다.
def square_of_even(numbers):
    result = []
    for num in numbers:
        if num % 2 == 0:
            result.append(num ** 2) # 이 줄에 return을 쓰면 함수가 종료됨
                                    # 반복을 하지 못함
    
    return result

############## 테스트 코드 삭제 금지 #################
print(square_of_even([1, 2, 3, 4, 5, 6]))  # [4, 16, 36]
#####################################################
```

```python
############## 주의 ##############
# 꼭 다시 보기 확인
# 딕셔너리에서 최고 점수 이름 찾기
# input 함수는 사용하지 않습니다.
# 여기에 코드를 작성하여 함수를 완성합니다.
def top_student(scores):
    max_score = -1        # 지금까지 본 최고 점수 (아무 점수보다 작은 값으로 시작)
    student = ""          # 최고 점수를 가진 학생 이름

    # 딕셔너리를 반복하면 기본적으로 'key(이름)'가 나온다
    for name in scores:
        score = scores[name]   # 이름으로 점수를 꺼냄

        # 현재 점수가 지금까지의 최고 점수보다 크면
        if score > max_score:
            max_score = score  # 최고 점수 갱신
            student = name     # 해당 학생 이름 저장

    return student              # 최고 점수의 학생 이름 반환


############## 테스트 코드 삭제 금지 #################
scores = {"kim": 85, "lee": 92, "park": 92, "choi": 78}
print(top_student(scores))  # lee
#####################################################
```
