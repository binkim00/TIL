```python
############## 주의 ##############
# 양수 개수 세기
# input 함수는 사용하지 않습니다.
def count_positive(numbers):
    # 여기에 코드를 작성하여 함수를 완성합니다.
    count = 0
    for num in numbers:
        if num > 0:
            count += 1
    return count


############## 테스트 코드 삭제 금지 #################
print(count_positive([1, -2, 3, 0, 5]))  # 3
print(count_positive([-1, -2, -3]))      # 0
#####################################################
```

```python
############## 주의 ##############
# 문자열 리스트에서 특정 길이 이상인 단어 개수 세기
# input 함수는 사용하지 않습니다.
def count_long_words(words, length):
    # 여기에 코드를 작성하여 함수를 완성합니다.
    count = 0
    for word in words:
        if len(word) >= length:
            count += 1
    return count


############## 테스트 코드 삭제 금지 #################
print(count_long_words(["hi", "hello", "bye"], 4))  # 1
print(count_long_words(["aa", "bb", "cc"], 2))      # 3
#####################################################
```

```python
############## 주의 ##############
# 숫자 리스트에서 가장 큰 값 찾기
# max 함수는 사용하지 않습니다.
# input 함수는 사용하지 않습니다.
def max_number(numbers):
    # 여기에 코드를 작성하여 함수를 완성합니다.
    max_num = 0
    for num in numbers:
        if num > max_num:
            max_num = num
    return max_num


############## 테스트 코드 삭제 금지 #################
print(max_number([3, 1, 7, 2]))  # 7
print(max_number([5, 5, 5]))     # 5
#####################################################
```

```python
############## 주의 ##############
# 문자열 리스트를 모두 소문자로 바꾼 리스트 반환
# input 함수는 사용하지 않습니다.
def to_lower(words):
    result = []
    for ch in words:
        result.append(ch.lower())  # 문자열을 소문자로 바꿔서 추가
    return result        

############## 테스트 코드 삭제 금지 #################
print(to_lower(["HELLO", "World"]))  # ['hello', 'world']
#####################################################
```

```python
############## 주의 ##############
# 딕셔너리에서 점수 평균 구하기
# input 함수는 사용하지 않습니다.
def average_score(scores):
    # 여기에 코드를 작성하여 함수를 완성합니다.
    total = 0
    students = 0

    for score in scores.values():
        students += 1
        total += score
        
    return total / students

# 짧게 내장 함수 쓰는 법
def average_score(scores):
    return sum(scores.values()) / len(scores)

############## 테스트 코드 삭제 금지 #################
scores = {"kim": 80, "lee": 90, "park": 70}
print(average_score(scores))  # 80.0
#####################################################
```

```python
############## 주의 ##############
# 문자열에서 특정 문자 개수 세기
# input 함수는 사용하지 않습니다.
def count_char(text, char):
    # 여기에 코드를 작성하여 함수를 완성합니다.
    count = 0
    for ch in text:
        if ch in char:
            count += 1
    return count


############## 테스트 코드 삭제 금지 #################
print(count_char("banana", "a"))  # 3
print(count_char("apple", "p"))   # 2
#####################################################
```

```python
############## 주의 ##############
# 리스트에서 기준값보다 큰 수만 담기
# input 함수는 사용하지 않습니다.
def filter_over(numbers, 기준값):
    # 여기에 코드를 작성하여 함수를 완성합니다.
    result = []
    for num in numbers:
        if num > 기준값:
            result.append(num)
    return result


############## 테스트 코드 삭제 금지 #################
print(filter_over([1, 5, 10, 3], 4))  # [5, 10]
#####################################################
```

```python
############## 주의 ##############
# 딕셔너리에서 가장 높은 점수의 학생 이름 찾기
# max 함수는 사용하지 않습니다.
# input 함수는 사용하지 않습니다.
def highest_student(scores):
    # 여기에 코드를 작성하여 함수를 완성합니다.
    max_score = 0
    hightes_student = ""
    for name, score in scores.items():
        if score > max_score:
            max_score = score
            hightes_student = name

    return hightes_student


############## 테스트 코드 삭제 금지 #################
scores = {"kim": 80, "lee": 95, "park": 70}
print(highest_student(scores))  # lee
#####################################################
```

```python
############## 주의 ##############
# 문자열 리스트에서 첫 글자만 모아 리스트로 반환
# input 함수는 사용하지 않습니다.
def first_letters(words):
    # 여기에 코드를 작성하여 함수를 완성합니다.
    result = []
    for ch in words:
        result.append(ch[0])
    return result


############## 테스트 코드 삭제 금지 #################
print(first_letters(["apple", "banana", "cherry"]))
# ['a', 'b', 'c']
#####################################################
```

```python
############## 주의 ##############
# 숫자 리스트에서 3의 배수만 문자열로 변환하여 반환
# input 함수는 사용하지 않습니다.
def three_multiple_to_string(numbers):
    # 여기에 코드를 작성하여 함수를 완성합니다.
    result = []
    for num in numbers:
        if num % 3 == 0:
            result.append(str(num))
            
    return result


############## 테스트 코드 삭제 금지 #################
print(three_multiple_to_string([1, 3, 6, 7, 9]))
# ['3', '6', '9']
#####################################################
```
