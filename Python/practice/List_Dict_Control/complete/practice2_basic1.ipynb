```python
############## 주의 ##############
# 성공^^
# 문자열 길이가 3 이상인 단어만 리스트로 반환
def filter_long_words(words):
    result = []

    for word in words:
        # 여기: 조건
        # 여기: 누적
        if len(word) >= 3:
            result.append(word)
            
    return result


############## 테스트 코드 삭제 금지 #################
print(filter_long_words(["hi", "hello", "a", "python"]))
# ['hello', 'python']
#####################################################
```

```python
############## 주의 ##############
# 한 번에 정답~
# 숫자 리스트에서 홀수의 합 구하기
def sum_odd(numbers):
    total = 0

    for num in numbers:
        # 여기: 조건
        # 여기: 누적
        if num % 2 != 0:
            total += num

    return total


############## 테스트 코드 삭제 금지 #################
print(sum_odd([1, 2, 3, 4, 5]))  # 9
#####################################################
```

```python
############## 주의 ##############
# 학생 이름만 리스트로 반환
def get_names(users):
    result = []

    for user in users:
        # 여기: 이름 꺼내기
        # 여기: 리스트에 추가
        name = user['name']
        result.append(name)

    return result


############## 테스트 코드 삭제 금지 #################
users = [
    {"name": "철수", "age": 20},
    {"name": "영희", "age": 22},
]
print(get_names(users))  # ['철수', '영희']
#####################################################
```

```python
############## 주의 ##############
# 이건 다시 봐야 할 듯
# 회사별 인원 수 세기
def count_by_company(users):
    result = {}

    for user in users:
        company = user["company"]

        # 여기: company가 result에 없을 때
        # 여기: 있을 때
        if company not in result:
            result[company] = 1
        else:
            result[company] += 1


    return result


############## 테스트 코드 삭제 금지 #################
users = [
    {"name": "철수", "company": "A"},
    {"name": "영희", "company": "B"},
    {"name": "민수", "company": "A"},
]
print(count_by_company(users))  # {'A': 2, 'B': 1}
#####################################################
```

```python
############## 주의 ##############
# 이것도 다시 한 번 보자
# 문자열에서 공백 제거 후 리스트로 반환
def split_words(text):
    # 여기: 문자열 쪼개기
    words = text.split()
    result = []

    for word in words:
        # 여기: 리스트에 추가
        result.append(word)

    return result


############## 테스트 코드 삭제 금지 #################
print(split_words("a b c"))  # ['a', 'b', 'c']
#####################################################
```

```python
############## 주의 ##############
# for문에 name, score 사용 후 조건 밑 리스트 추가 학습 필요
# 점수가 80점 이상인 학생 이름만 반환
def high_score_students(scores):
    result = []

    for name, score in scores.items():
        # 여기: 조건
        # 여기: 리스트에 추가
        if score >= 80:
            result.append(name)

    return result


############## 테스트 코드 삭제 금지 #################
scores = {"kim": 80, "lee": 95, "park": 70}
print(high_score_students(scores))  # ['kim', 'lee']
#####################################################
```

```python
############## 주의 ##############
# 이건 Easy
# 짝수만 2배해서 리스트로 반환
def double_even(numbers):
    result = []

    for num in numbers:
        # 여기: 조건
        # 여기: 가공해서 추가
        if num % 2 == 0:
            result.append(num * 2)

    return result


############## 테스트 코드 삭제 금지 #################
print(double_even([1, 2, 3, 4, 5]))  # [4, 8]
#####################################################
```
