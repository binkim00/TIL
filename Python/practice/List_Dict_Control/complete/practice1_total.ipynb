```python
############## 주의 ##############
# input 함수는 사용하지 않습니다.
#
# [문제]
# 문자열 text에서 문자 target이 몇 번 등장하는지 반환하시오.
#
# [요구조건]
# - 문자열을 반복하여 하나씩 비교한다
def count_char(text, target):
    # 여기에 코드를 작성하여 함수를 완성합니다.
    r = 0
    for ch in text:
        if target in ch:
            r += 1
    
    return r


############## 테스트 코드 삭제 금지 #################
print(count_char("banana", "a"))  # 3
print(count_char("hello", "z"))   # 0
#####################################################
```

```python
############## 주의 ##############
# input 함수는 사용하지 않습니다.
#
# [문제]
# 2차원 리스트 matrix에 들어있는 모든 숫자의 합을 반환하시오.
#
# [요구조건]
# - 중첩 반복문을 사용한다
def sum_matrix(matrix):
    # 여기에 코드를 작성하여 함수를 완성합니다.
    r = 0
    for list in matrix:

        for num in list:
                r += num
    
    return r


############## 테스트 코드 삭제 금지 #################
print(sum_matrix([[1, 2], [3, 4]]))        # 10
print(sum_matrix([[5, 5, 5], [1, 1, 1]]))  # 18
#####################################################
```

```python
############## 주의 ##############
# input 함수는 사용하지 않습니다.
#
# [문제]
# 문장 sentence에서 공백 기준으로 단어를 나눈 후
# 가장 많이 등장한 단어를 반환하시오.
# (동점이면 먼저 등장한 단어를 반환)
#
# [요구조건]
# - 단어 빈도를 딕셔너리로 센다
# - 최대 빈도를 가진 단어를 찾는다
def most_common_word(sentence):
    # 1. 문장을 공백 기준으로 나눠 리스트로 만든다
    words = sentence.split()
    
    # 2. 단어 빈도를 저장할 딕셔너리
    count_dict = {}
    
    # 3. 각 단어의 등장 횟수 세기
    for word in words:
        if word in count_dict:
            count_dict[word] += 1
        else:
            count_dict[word] = 1

    # 4. 가장 많이 등장한 단어 찾기
    max_count = 0
    result = None

    for word in words:
        # words를 기준으로 순회 → 먼저 나온 단어 우선
        if count_dict[word] > max_count:
            max_count = count_dict[word]
            result = word

    return result

############## 테스트 코드 삭제 금지 #################
print(most_common_word("apple banana apple apple banana"))  # apple
print(most_common_word("hi hi hello hello"))                # hi
#####################################################
```

```python
############## 주의 ##############
# input 함수는 사용하지 않습니다.
#
# [문제]
# numbers 리스트에서
# 같은 값이 연속으로 등장한 최대 길이를 반환하시오.
#
# [예시]
# [1,1,2,2,2,3] → 3
#
# [요구조건]
# - 이전 값과 현재 값을 비교한다
# - 연속이 끊기면 길이를 초기화한다
def longest_same(numbers):
    # 첫 값 기준 세팅
    prev = numbers[0]
    count = 1
    max_count = 1

    # 두 번째 값부터 비교
    for num in numbers[1:]:
        if num == prev:
            count += 1
        else:
            count = 1  # 연속 끊김 → 초기화

        if count > max_count:
            max_count = count

        prev = num  # 이전 값 갱신

    return max_count

############## 테스트 코드 삭제 금지 #################
print(longest_same([1, 1, 2, 2, 2, 3]))  # 3
print(longest_same([5, 5, 5, 5]))        # 4
#####################################################
```

```python
############## 주의 ##############
# input 함수는 사용하지 않습니다.
#
# [문제]
# numbers 리스트에서
# "앞의 값 < 현재 값 > 다음 값" 패턴의 개수를 반환하시오.
#
# [예시]
# [1, 3, 2, 4, 1] → 2 (3, 4)
#
# [요구조건]
# - 현재 값 기준으로 양쪽 값을 비교한다
def count_peaks(numbers):
    # 1. 리스트 길이 직접 구하기
    length = 0
    for _ in numbers:
        length += 1

    count = 0
    i = 1  # 두 번째 요소부터 시작

    # 2. 앞 < 현재 > 뒤 비교
    while i < length - 1:
        if numbers[i - 1] < numbers[i] and numbers[i] > numbers[i + 1]:
            count += 1
        i += 1

    return count

############## 테스트 코드 삭제 금지 #################
print(count_peaks([1, 3, 2, 4, 1]))  # 2
print(count_peaks([1, 2, 3, 4]))     # 0
#####################################################
```

```python
############## 주의 ##############
# input 함수는 사용하지 않습니다.
#
# [문제]
# students 리스트에는 학생 정보가 들어 있다.
# score가 60 이상이면 "pass", 미만이면 "fail"로 분류하여
# 결과를 딕셔너리로 반환하시오.
#
# [요구조건]
# - 이름을 key, 결과를 value로 저장한다
def classify_students(students):
    r = {}

    for student in students:
        name = student["name"]
        score = student["score"]

        if score >= 60:
            r[name] = "pass"
        else:
            r[name] = "fail"

    return r

############## 테스트 코드 삭제 금지 #################
students = [
    {"name": "kim", "score": 85},
    {"name": "lee", "score": 55},
    {"name": "park", "score": 70},
]
print(classify_students(students))
# {'kim': 'pass', 'lee': 'fail', 'park': 'pass'}
#####################################################
```
