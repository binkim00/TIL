```python
############## 주의 ##############
# 복잡한 JSON 구조에서 데이터 수집/가공/출력 연습
# input 함수는 사용하지 않습니다.

data = {
    "meta": {"count": 4},
    "results": [
        {
            "id": 1,
            "user": {"name": "Kim", "email": "kim@test.com"},
            "company": {"name": "A-Corp", "location": {"city": "Seoul", "geo": {"lat": 37.56, "lng": 126.97}}},
            "tags": ["backend", "python"],
            "scores": [80, 90, 70],
        },
        {
            "id": 2,
            "user": {"name": "Lee", "email": "lee@test.com"},
            "company": {"name": "B-Lab", "location": {"city": "Busan", "geo": {"lat": 35.18, "lng": 129.07}}},
            "tags": ["frontend"],
            "scores": [60, 75],
        },
        {
            "id": 3,
            "user": {"name": "Park", "email": "park@test.com"},
            "company": {"name": "A-Corp", "location": {"city": "Seoul", "geo": {"lat": 37.50, "lng": 127.02}}},
            "tags": [],
            "scores": [100],
        },
        {
            "id": 4,
            "user": {"name": "Choi", "email": "choi@test.com"},
            "company": {"name": "C-Works", "location": {"city": "Incheon", "geo": {"lat": 37.45, "lng": 126.70}}},
            "tags": ["python", "data"],
            "scores": [88, 92, 95, 85],
        },
    ]
}

############## 주의 ##############
# 4) 회사별 사용자 이름 모으기 (딕셔너리 누적)
# 결과 예시: {'A-Corp': ['Kim', 'Park'], 'B-Lab': ['Lee'], 'C-Works': ['Choi']}
# input 함수는 사용하지 않습니다.
def group_users_by_company(data):
    group_list = {}  # {'회사명': [이름들]}

    for result in data["results"]:
        company_name = result["company"]["name"]  # 회사명
        user_name = result["user"]["name"]        # 사용자 이름

        if company_name not in group_list:        # 회사 key가 없으면
            group_list[company_name] = []         # 빈 리스트로 생성

        group_list[company_name].append(user_name) # 해당 회사 리스트에 이름 추가

    return group_list


############## 테스트 코드 삭제 금지 #################
print(group_users_by_company(data))
# {'A-Corp': ['Kim', 'Park'], 'B-Lab': ['Lee'], 'C-Works': ['Choi']}
#####################################################
```

```python
############## 주의 ##############
# 복잡한 JSON 구조에서 데이터 수집/가공/출력 연습
# input 함수는 사용하지 않습니다.

data = {
    "meta": {"count": 4},
    "results": [
        {
            "id": 1,
            "user": {"name": "Kim", "email": "kim@test.com"},
            "company": {"name": "A-Corp", "location": {"city": "Seoul", "geo": {"lat": 37.56, "lng": 126.97}}},
            "tags": ["backend", "python"],
            "scores": [80, 90, 70],
        },
        {
            "id": 2,
            "user": {"name": "Lee", "email": "lee@test.com"},
            "company": {"name": "B-Lab", "location": {"city": "Busan", "geo": {"lat": 35.18, "lng": 129.07}}},
            "tags": ["frontend"],
            "scores": [60, 75],
        },
        {
            "id": 3,
            "user": {"name": "Park", "email": "park@test.com"},
            "company": {"name": "A-Corp", "location": {"city": "Seoul", "geo": {"lat": 37.50, "lng": 127.02}}},
            "tags": [],
            "scores": [100],
        },
        {
            "id": 4,
            "user": {"name": "Choi", "email": "choi@test.com"},
            "company": {"name": "C-Works", "location": {"city": "Incheon", "geo": {"lat": 37.45, "lng": 126.70}}},
            "tags": ["python", "data"],
            "scores": [88, 92, 95, 85],
        },
    ]
}

############## 주의 ##############
# 5) 사용자별 점수 평균 딕셔너리 만들기
# 각 결과에서 user["name"]을 key로, scores 평균을 value로 저장한 딕셔너리를 반환합니다.
# 평균은 float로 반환되도록 합니다.
# input 함수는 사용하지 않습니다.
def user_score_avg(data):
    avg_dict = {}  # {'이름': 평균}

    for result in data["results"]:
        name = result["user"]["name"]
        scores = result["scores"]

        total = 0
        count = 0
        for s in scores:
            total += s
            count += 1

        avg_dict[name] = total / count  # 평균(실수)

    return avg_dict


############## 테스트 코드 삭제 금지 #################
print(user_score_avg(data))
# {'Kim': 80.0, 'Lee': 67.5, 'Park': 100.0, 'Choi': 90.0}
#####################################################
```

```python
############## 주의 ##############
# 복잡한 JSON 구조에서 데이터 수집/가공/출력 연습
# input 함수는 사용하지 않습니다.

data = {
    "meta": {"count": 4},
    "results": [
        {
            "id": 1,
            "user": {"name": "Kim", "email": "kim@test.com"},
            "company": {"name": "A-Corp", "location": {"city": "Seoul", "geo": {"lat": 37.56, "lng": 126.97}}},
            "tags": ["backend", "python"],
            "scores": [80, 90, 70],
        },
        {
            "id": 2,
            "user": {"name": "Lee", "email": "lee@test.com"},
            "company": {"name": "B-Lab", "location": {"city": "Busan", "geo": {"lat": 35.18, "lng": 129.07}}},
            "tags": ["frontend"],
            "scores": [60, 75],
        },
        {
            "id": 3,
            "user": {"name": "Park", "email": "park@test.com"},
            "company": {"name": "A-Corp", "location": {"city": "Seoul", "geo": {"lat": 37.50, "lng": 127.02}}},
            "tags": [],
            "scores": [100],
        },
        {
            "id": 4,
            "user": {"name": "Choi", "email": "choi@test.com"},
            "company": {"name": "C-Works", "location": {"city": "Incheon", "geo": {"lat": 37.45, "lng": 126.70}}},
            "tags": ["python", "data"],
            "scores": [88, 92, 95, 85],
        },
    ]
}

############## 주의 ##############
# 6) 좌표 범위 필터링 후 가공 리스트 반환
# lat, lng가 모두 -80 ~ 80 범위 안인 경우만 남기고,
# 각 항목을 {"name": 이름, "lat": lat, "lng": lng, "company": 회사명} 형태로 리스트에 담아 반환합니다.
# input 함수는 사용하지 않습니다.
def filter_geo_range(data):
    filtered = []

    for result in data["results"]:
        name = result["user"]["name"]
        company = result["company"]["name"]

        geo = result["company"]["location"]["geo"]
        lat = geo["lat"]
        lng = geo["lng"]

        if -80 <= lat <= 80 and -80 <= lng <= 80:
            filtered.append({
                "name": name,
                "lat": lat,
                "lng": lng,
                "company": company
            })

    return filtered


############## 테스트 코드 삭제 금지 #################
print(filter_geo_range(data))
# [
#   {'name': 'Kim', 'lat': 37.56, 'lng': 126.97, 'company': 'A-Corp'},
#   {'name': 'Lee', 'lat': 35.18, 'lng': 129.07, 'company': 'B-Lab'},
#   {'name': 'Park', 'lat': 37.5, 'lng': 127.02, 'company': 'A-Corp'},
#   {'name': 'Choi', 'lat': 37.45, 'lng': 126.7, 'company': 'C-Works'}
# ]
#####################################################
```

```python
############## 주의 ##############
# 복잡한 JSON 구조에서 데이터 수집/가공/출력 연습
# input 함수는 사용하지 않습니다.

data = {
    "meta": {"count": 4},
    "results": [
        {
            "id": 1,
            "user": {"name": "Kim", "email": "kim@test.com"},
            "company": {"name": "A-Corp", "location": {"city": "Seoul", "geo": {"lat": 37.56, "lng": 126.97}}},
            "tags": ["backend", "python"],
            "scores": [80, 90, 70],
        },
        {
            "id": 2,
            "user": {"name": "Lee", "email": "lee@test.com"},
            "company": {"name": "B-Lab", "location": {"city": "Busan", "geo": {"lat": 35.18, "lng": 129.07}}},
            "tags": ["frontend"],
            "scores": [60, 75],
        },
        {
            "id": 3,
            "user": {"name": "Park", "email": "park@test.com"},
            "company": {"name": "A-Corp", "location": {"city": "Seoul", "geo": {"lat": 37.50, "lng": 127.02}}},
            "tags": [],
            "scores": [100],
        },
        {
            "id": 4,
            "user": {"name": "Choi", "email": "choi@test.com"},
            "company": {"name": "C-Works", "location": {"city": "Incheon", "geo": {"lat": 37.45, "lng": 126.70}}},
            "tags": ["python", "data"],
            "scores": [88, 92, 95, 85],
        },
    ]
}

############## 주의 ##############
# 7) 출력 문자열 만들기 (가공 + 출력용 문자열 리스트)
# 각 사용자에 대해 아래 형식의 문자열을 만들어 리스트로 반환합니다.
# 형식: "{이름}({회사명}) - {도시} - tags:{태그개수}"
# input 함수는 사용하지 않습니다.
def make_summary_lines(data):
    lines = []

    for result in data["results"]:
        name = result["user"]["name"]
        company = result["company"]["name"]
        city = result["company"]["location"]["city"]

        # tags 개수 세기 (len 안 쓰고 카운트)
        tag_count = 0
        for _ in result["tags"]:
            tag_count += 1

        lines.append(f"{name}({company}) - {city} - tags:{tag_count}")

    return lines


############## 테스트 코드 삭제 금지 #################
print(make_summary_lines(data))
# [
#   'Kim(A-Corp) - Seoul - tags:2',
#   'Lee(B-Lab) - Busan - tags:1',
#   'Park(A-Corp) - Seoul - tags:0',
#   'Choi(C-Works) - Incheon - tags:2'
# ]
#####################################################
```
