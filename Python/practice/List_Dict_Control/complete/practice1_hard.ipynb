```python
############## 주의 ##############
# input 함수는 사용하지 않습니다.
#
# [문제]
# numbers 리스트에서 "이전 값보다 커진 경우"가 총 몇 번 발생했는지 반환하시오.
# 즉, i가 1부터 끝까지일 때,
# numbers[i] > numbers[i-1] 인 경우의 개수를 세어 반환합니다.
def count_increases(numbers):
    result = 0
    prev = numbers[0]           # 이전 값 저장

    for num in numbers[1:]:     # 두 번째 값부터 비교
        if num > prev:          # 현재 값이 이전 값보다 크면
            result += 1         # 증가 횟수 1 증가
        prev = num              # 이전 값 갱신

    return result


############## 테스트 코드 삭제 금지 #################
print(count_increases([1, 2, 3, 1, 2]))  # 3
print(count_increases([5, 4, 3, 2, 1]))  # 0
print(count_increases([1, 2, 3, 4, 5]))  # 4
#####################################################
```

```python
############## 주의 ##############
# input 함수는 사용하지 않습니다.
#
# [문제]
# numbers 리스트에서 "연속으로 증가하는 구간"의 최대 길이를 반환하시오.
# 단, "증가가 최소 1번 이상 발생한 구간"만 구간으로 인정합니다.
# 증가가 한 번도 없으면 0을 반환합니다.
#
# [요구조건]
# - 숫자가 이전 값보다 클 때만 증가로 판단 (num > prev)
# - 증가가 끊기면 현재 길이를 1로 초기화
# - 증가가 한 번도 없으면 0 반환
def longest_increasing_B(numbers):
    # 예외: 빈 리스트면 증가 구간이 있을 수 없으므로 0
    if not numbers:
        return 0

    max_len = 1         # (임시) 최대 길이 기록, 최소 1부터 시작
    current_len = 1     # 현재 연속 증가 구간 길이 (원소 개수)
    prev = numbers[0]   # 이전 값 저장
    has_increase = False  # 증가가 실제로 한 번이라도 있었는지 체크

    # 두 번째 원소부터 비교 시작
    for num in numbers[1:]:
        if num > prev:              # 증가라면
            current_len += 1        # 현재 구간 길이 증가
            has_increase = True     # 증가가 있었다고 표시
        else:                       # 증가가 끊겼다면(같거나 작아졌다면)
            current_len = 1         # 현재 구간 길이 초기화(새 구간 시작)

        # 최대 길이 갱신
        if current_len > max_len:
            max_len = current_len

        prev = num                  # 다음 비교를 위해 이전 값 갱신

    # 증가가 한 번도 없었다면(모든 구간이 길이 1뿐) 0 반환
    if not has_increase:
        return 0

    return max_len


############## 테스트 코드 삭제 금지 #################
print(longest_increasing_B([1, 2, 3, 1, 2]))  # 3
print(longest_increasing_B([5, 4, 3, 2, 1]))  # 0  (증가가 한 번도 없음)
print(longest_increasing_B([1, 2, 3, 4, 5]))  # 5
print(longest_increasing_B([2, 2, 2]))        # 0  (커진 적 없음)
#####################################################
```

```python
############## 주의 ##############
# input 함수는 사용하지 않습니다.
#
# [문제]
# numbers 리스트에서
# "짝수이면서 10보다 큰 최초의 값"을 반환하시오.
# 없으면 -1을 반환하시오.
#
# [요구조건]
# - 조건을 만족하면 즉시 종료
# - 끝까지 없으면 -1 반환
def first_even_over_ten(numbers):
    # 여기에 코드를 작성하여 함수를 완성합니다.
    for num in numbers:                 # 리스트 순회
        if num % 2 == 0 and num > 10:   # 조건
            return num                  # 조건 만족하는 값 즉시 반환
        
    return -1                           # 끝까지 없으면 -1 반환


############## 테스트 코드 삭제 금지 #################
print(first_even_over_ten([1, 3, 8, 9, 12, 14]))  # 12
print(first_even_over_ten([2, 4, 6, 8]))         # -1
#####################################################
```

```python
############## 주의 ##############
# input 함수는 사용하지 않습니다.
#
# [문제]
# scores 딕셔너리에서
# 점수가 기준점수 이상인 사람의 이름만
# 리스트로 반환하시오.
#
# [요구조건]
# - 반환 순서는 딕셔너리 순서를 따른다
def passed_students(scores, 기준점수):
    # 여기에 코드를 작성하여 함수를 완성합니다.
    result = []

    for name, score in scores.items():
        
        if score >= 기준점수:
            result.append(name)
    
    return result


############## 테스트 코드 삭제 금지 #################
scores = {"kim": 85, "lee": 55, "park": 70}
print(passed_students(scores, 60))  # ['kim', 'park']
#####################################################
```

```python
############## 주의 ##############
# input 함수는 사용하지 않습니다.
#
# text 문자열에서 가장 많이 등장한 문자를 반환
# (공백도 문자로 취급)
def most_common_char(text):
    counts = {}                      # 문자별 등장 횟수를 저장할 딕셔너리 생성

    for ch in text:                  # 문자열을 한 글자씩 순회
        if ch in counts:             # 이미 등장한 문자라면
            counts[ch] += 1          # 등장 횟수를 1 증가
        else:                        # 처음 등장한 문자라면
            counts[ch] = 1           # 등장 횟수를 1로 초기화

    best_char = text[0]              # 가장 많이 등장한 문자 기준값(첫 문자)
    best_count = counts[best_char]   # 기준 문자의 등장 횟수 저장

    for ch in text:                  # 문자열을 다시 처음부터 순회
        if counts[ch] > best_count:  # 현재 문자의 등장 횟수가 더 많다면
            best_char = ch           # 기준 문자를 현재 문자로 변경
            best_count = counts[ch]  # 기준 등장 횟수도 함께 갱신

    return best_char                 # 최종적으로 가장 많이 등장한 문자 반환


############## 테스트 코드 삭제 금지 #################
print(most_common_char("jazz is good"))  # 'z'
print(most_common_char("aabbcc"))        # 'a'
#####################################################
```

```python
############## 주의 ##############
# input 함수는 사용하지 않습니다.
#
# [문제]
# orders 리스트에는 주문 상태가 순서대로 기록되어 있다.
# "PAID" 다음에 바로 "CANCEL"이 나온 횟수를 구하시오.
#
# [예시]
# ["PAID", "CANCEL", "PAID", "DONE"] → 1
#
# [요구조건]
# - 현재 상태와 다음 상태를 함께 비교한다
# 여기에 코드를 작성하여 함수를 완성합니다.
def count_paid_cancel(orders):
    result = 0

    for i in range(len(orders) - 1):            # 마지막 전까지만 순회
        if orders[i] == "PAID" and orders[i+1] == "CANCEL":
            result += 1                          # 조건 만족 시 카운트

    return result

# 내장 함수 제거
def count_paid_cancel(orders):
    result = 0
    prev = None                       # 이전 상태(현재 비교 기준)

    for current in orders:            # 리스트를 앞에서부터 순회
        if prev == "PAID" and current == "CANCEL":  # (이전,현재) 쌍이 PAID→CANCEL이면
            result += 1               # 횟수 1 증가
        prev = current                # 다음 비교를 위해 이전 상태 갱신

    return result


############## 테스트 코드 삭제 금지 #################
print(count_paid_cancel(["PAID", "CANCEL", "PAID", "DONE"]))      # 1
print(count_paid_cancel(["PAID", "PAID", "CANCEL", "CANCEL"]))   # 1
#####################################################
```
