```python
############## 주의 ##############
# 홀수 개수 세기
# input 함수는 사용하지 않습니다.
def count_odd(numbers):
    # 여기에 코드를 작성하여 함수를 완성합니다.
    count = 0
    for num in numbers:
        if num % 2 != 0:
            count += 1
    return count


############## 테스트 코드 삭제 금지 #################
print(count_odd([1, 2, 3, 4, 6, 7]))  # 3
print(count_odd([2, 4, 6, 8]))        # 0
print(count_odd([1, 3, 5]))           # 3
#####################################################
```

```python
############## 주의 ##############
# 리스트 평균 구하기
# input 함수는 사용하지 않습니다.
def average(numbers):
    # 여기에 코드를 작성하여 함수를 완성합니다.
    total = 0
    count = 0
    for num in numbers:
        total += num
        count += 1
    
    return total / count


############## 테스트 코드 삭제 금지 #################
print(average([10, 20, 30, 40]))  # 25.0
print(average([5, 5, 5, 5]))      # 5.0
#####################################################
```

```python
############## 주의 ##############
# 기준값보다 작은 값 개수 세기
# input 함수는 사용하지 않습니다.
def count_under(numbers, 기준값):
    # 여기에 코드를 작성하여 함수를 완성합니다.
    r = 0
    for n in numbers:
        if n < 기준값:
            r += 1

    return r


############## 테스트 코드 삭제 금지 #################
print(count_under([10, 20, 30, 40], 25))  # 2
print(count_under([5, 5, 5, 5], 5))       # 0
#####################################################
```

```python
############## 주의 ##############
# 다시 보자
# 가장 긴 문자열 찾기
# input 함수는 사용하지 않습니다.
def longest_word(words):
    longest = words[0]          # 기준값 설정 / 여기 중요

    for word in words:
        if len(word) > len(longest):
            longest = word      # 더 길면 기준 교체

    return longest


############## 테스트 코드 삭제 금지 #################
print(longest_word(["hi", "hello", "a"]))     # hello
print(longest_word(["aa", "bb", "c"]))        # aa
#####################################################
```

```python
############## 주의 ##############
# 다시 보자
# 가장 긴 문자열 찾기 (len 사용 금지)
# input 함수는 사용하지 않습니다.
def longest_word(words):
    longest = words[0] # 여기 중요

    for word in words:
        # word 길이 세기 / 각 단어 길이 세기
        word_len = 0
        for _ in word:
            word_len += 1

        # longest 길이 세기 / 가장 긴 단어 길이 세기
        longest_len = 0
        for _ in longest:
            longest_len += 1

        # 길이 비교 / 비교 후 더 긴 단어 넣기
        if word_len > longest_len:
            longest = word

    return longest


############## 테스트 코드 삭제 금지 #################
print(longest_word(["hi", "hello", "a"]))     # hello
print(longest_word(["aa", "bb", "c"]))        # aa
#####################################################
```

```python
############## 주의 ##############
# 딕셔너리 값의 합 구하기
# input 함수는 사용하지 않습니다.
def sum_scores(scores):
    # 여기에 코드를 작성하여 함수를 완성합니다.
    r = 0
    for score in scores.values():
        r += score        
    return r


############## 테스트 코드 삭제 금지 #################
scores = {"kim": 80, "lee": 90, "park": 70}
print(sum_scores(scores))  # 240
#####################################################
```

```python
############## 주의 ##############
# 가장 낮은 점수의 학생 이름 찾기
# input 함수는 사용하지 않습니다.
# min 함수는 사용하지 않습니다.
def lowest_student(scores):      # scores라는 딕셔너리를 받아서 최저 점수 학생 이름을 반환하는 함수
    first = True                 # 첫 번째 데이터를 만났는지 확인하는 표시(초기 기준값 설정용)

    for name, score in scores.items():   # 딕셔너리에서 (키=이름, 값=점수) 쌍을 하나씩 꺼내 반복
        if first:                        # 지금이 첫 번째 반복이라면(아직 기준값이 없다면)
            min_score = score            # 현재 점수를 '최소 점수' 기준값으로 저장
            min_student = name           # 현재 이름을 '최소 점수 학생' 기준값으로 저장
            first = False                # 이제 기준값을 만들었으니, 다음부터는 첫 번째가 아님으로 표시
        else:                            # 두 번째 데이터부터는(기준값이 이미 있으니)
            if score < min_score:        # 현재 점수가 지금까지의 최소 점수보다 더 작다면
                min_score = score        # 최소 점수를 현재 점수로 갱신(교체)
                min_student = name       # 최소 점수 학생 이름도 현재 이름으로 갱신(교체)

    return min_student                   # 반복이 끝나면 최종 최소 점수 학생 이름을 반환


############## 테스트 코드 삭제 금지 #################
scores = {"kim": 85, "lee": 92, "park": 78}  # 이름:점수 형태의 딕셔너리 생성
print(lowest_student(scores))  # park        # 함수 실행 결과(최저 점수 78인 park이 출력되어야 함)
#####################################################
```

```python
############## 주의 ##############
# 조건 필터 + 가공
# 홀수만 2배해서 리스트에 담기
# input 함수는 사용하지 않습니다.
def double_odd(numbers):
    # 여기에 코드를 작성하여 함수를 완성합니다.
    r = []
    for n in numbers:
        if n % 2 != 0:
            r.append(n * 2)
    return r


############## 테스트 코드 삭제 금지 #################
print(double_odd([1, 2, 3, 4, 5]))  # [2, 6, 10]
#####################################################
```
