```python
############## 주의 ##############
# input 함수는 사용하지 않습니다.
# 겨우 맞춤
# [문제]
# numbers 리스트에서 "평균 이상"인 값의 개수를 반환하시오.
#
# [요구조건]
# 1. 리스트의 평균을 먼저 구한다
# 2. 평균 이상인 값을 센다
def count_over_avg(numbers):
    # 여기에 코드를 작성하여 함수를 완성합니다.
    total = 0
    count = 0
    
    for num in numbers:
        total += num
        count += 1

    avg = total / count
    r = 0
    
    for num in numbers:
        if num >= avg:
            r += 1

    return r


############## 테스트 코드 삭제 금지 #################
print(count_over_avg([10, 20, 30, 40]))  # 2
print(count_over_avg([5, 5, 5, 5]))      # 4
#####################################################
```

```python
############## 주의 ##############
# input 함수는 사용하지 않습니다.
#
# [문제]
# numbers 리스트에서 중복을 제거한 새 리스트를 반환하시오.
# 단, 기존 순서는 유지해야 합니다.
#
# [요구조건]
# - 이미 나온 값은 다시 추가하지 않는다
def remove_duplicates(numbers):
    # 여기에 코드를 작성하여 함수를 완성합니다.
    result = []

    for num in numbers:
        if num not in result:
            result.append(num)
        
    return result


############## 테스트 코드 삭제 금지 #################
print(remove_duplicates([1, 2, 2, 3, 3, 3, 4]))  # [1, 2, 3, 4]
print(remove_duplicates([5, 5, 5]))              # [5]
#####################################################
```

```python
############## 주의 ##############
# input 함수는 사용하지 않습니다.
#
# [문제]
# numbers 리스트에서
# "짝수이면서 5보다 큰 값"들의 합을 반환하시오.
#
# [요구조건]
# - 조건을 동시에 만족해야 한다
def sum_even_over_five(numbers):
    # 여기에 코드를 작성하여 함수를 완성합니다.
    r = 0
    for num in numbers:
        if num % 2 == 0 and num > 5:
            r += num
            
    return r

############## 테스트 코드 삭제 금지 #################
print(sum_even_over_five([1, 2, 4, 6, 8, 10]))  # 24
print(sum_even_over_five([2, 4]))               # 0
#####################################################
```

```python
############## 주의 ##############
# input 함수는 사용하지 않습니다.
#
# [문제]
# scores 딕셔너리에 들어있는 점수들의 평균을 반환하시오.
#
# [요구조건]
# - 딕셔너리를 반복하여 값을 누적한다
def average_score(scores):
    # 여기에 코드를 작성하여 함수를 완성합니다.
    count = 0
    total = 0    
    for score in scores.values():
        count += 1
        total += score
    
    avg = total / count

    return avg


############## 테스트 코드 삭제 금지 #################
scores = {"kim": 80, "lee": 90, "park": 70}
print(average_score(scores))  # 80.0
#####################################################
```

```python
############## 주의 ##############
# input 함수는 사용하지 않습니다.
#
# [문제]
# words 리스트에서 가장 길이가 긴 문자열을 반환하시오.
# 길이가 같으면 먼저 나온 문자열을 유지합니다.
#
# [요구조건]
# - 문자열 길이를 기준으로 비교한다
def longest_word(words):
    # 여기에 코드를 작성하여 함수를 완성합니다.
    longest = words[0] # 단어 하나 주입
    
    for word in words:
        # 단어 길이 변수 할당
        word_len = 0

        # 단어 길이 세기
        for ch in word:
            word_len += 1

        # 최장 단어 길이 변수 할당
        longest_len = 0
        
        # 최장 단어 길이 세기
        for ch in longest:
            longest_len += 1
        
        # 단어 길이와 최장 단어 길이 비교
        if word_len > longest_len:
            longest = word
        
    return longest


############## 테스트 코드 삭제 금지 #################
print(longest_word(["hi", "hello", "hey"]))     # hello
print(longest_word(["a", "bb", "cc"]))          # bb
#####################################################
```

```python
############## 주의 ##############
# input 함수는 사용하지 않습니다.
#
# [문제]
# numbers 리스트에서
# "앞의 값보다 커진 경우"가 몇 번 발생했는지 반환하시오.
#
# [예시]
# [1, 3, 2, 4] → 2 (1→3, 2→4)
#
# [요구조건]
# - 현재 값과 이전 값을 비교한다
def count_increase(numbers):
    # 여기에 코드를 작성하여 함수를 완성합니다.
    result = 0 # 결과 저장할 변수

    xnum = numbers[0] # 첫 값부터 시작 / 첫 값을 이전 값으로 저장

    for num in numbers[1:]: # 두 번째 값부터 비교 반복 시작
        if num > xnum:      # 현재 값이 이전 값보다 크면
            result += 1     # 결과 값 + 1

        xnum = num          # 이전 값 갱신

    return result


############## 테스트 코드 삭제 금지 #################
print(count_increase([1, 3, 2, 4]))     # 2
print(count_increase([5, 4, 3, 2]))     # 0
#####################################################
```
