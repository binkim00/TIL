```python
############## 주의 ##############
# input 함수는 사용하지 않습니다.
#
# [문제]
# words 리스트에서 가장 길이가 긴 문자열을 반환하시오.
# 길이가 같으면 먼저 나온 문자열을 유지합니다.
#
# [요구조건]
# - 문자열 길이를 기준으로 비교한다
def longest_word(words):
    # 여기에 코드를 작성하여 함수를 완성합니다.
    longest = words[0] # 단어 하나 주입
    
    for word in words:
        # 단어 길이 변수 할당
        word_len = 0

        # 단어 길이 세기
        for ch in word:
            word_len += 1

        # 최장 단어 길이 변수 할당
        longest_len = 0
        
        # 최장 단어 길이 세기
        for ch in longest:
            longest_len += 1
        
        # 단어 길이와 최장 단어 길이 비교
        if word_len > longest_len:
            longest = word
        
    return longest


############## 테스트 코드 삭제 금지 #################
print(longest_word(["hi", "hello", "hey"]))     # hello
print(longest_word(["a", "bb", "cc"]))          # bb
#####################################################
```

```python
############## 주의 ##############
# input 함수는 사용하지 않습니다.
#
# [문제]
# 문장 sentence에서 공백 기준으로 단어를 나눈 후
# 가장 많이 등장한 단어를 반환하시오.
# (동점이면 먼저 등장한 단어를 반환)
#
# [요구조건]
# - 단어 빈도를 딕셔너리로 센다
# - 최대 빈도를 가진 단어를 찾는다
def most_common_word(sentence):
    # 1. 문장을 공백 기준으로 나눠 리스트로 만든다
    words = sentence.split()
    
    # 2. 단어 빈도를 저장할 딕셔너리
    count_dict = {}
    
    # 3. 각 단어의 등장 횟수 세기
    for word in words:
        if word in count_dict:
            count_dict[word] += 1
        else:
            count_dict[word] = 1

    # 4. 가장 많이 등장한 단어 찾기
    max_count = 0
    result = None

    for word in words:
        # words를 기준으로 순회 → 먼저 나온 단어 우선
        if count_dict[word] > max_count:
            max_count = count_dict[word]
            result = word

    return result

############## 테스트 코드 삭제 금지 #################
print(most_common_word("apple banana apple apple banana"))  # apple
print(most_common_word("hi hi hello hello"))                # hi
#####################################################
```

```python
############## 주의 ##############
# input 함수는 사용하지 않습니다.
#
# [문제]
# scores 딕셔너리에 들어있는 점수들의 평균을 반환하시오.
#
# [요구조건]
# - 딕셔너리를 반복하여 값을 누적한다
def average_score(scores):
    # 여기에 코드를 작성하여 함수를 완성합니다.
    count = 0
    total = 0    
    for score in scores.values():
        count += 1
        total += score
    
    avg = total / count

    return avg


############## 테스트 코드 삭제 금지 #################
scores = {"kim": 80, "lee": 90, "park": 70}
print(average_score(scores))  # 80.0
#####################################################
```

```python
############## 주의 ##############
# 완성 못함
# 가장 많이 등장한 값 찾기
# input 함수는 사용하지 않습니다.
# 여기에 코드를 작성하여 함수를 완성합니다.
def find_mode(numbers):
    counts = {}          
    # 각 숫자가 몇 번 나왔는지 저장할 딕셔너리
    # 예: {2: 3} → 숫자 2가 3번 나옴

    max_count = 0        
    # 지금까지 확인한 숫자 중 가장 많이 나온 횟수

    mode = numbers[0]    
    # 가장 많이 나온 숫자를 임시로 저장
    # 일단 첫 번째 숫자를 기준으로 잡음

    # 1단계: 숫자별 등장 횟수 세기
    for num in numbers:      
        # numbers 리스트에서 숫자를 하나씩 꺼냄

        if num in counts:    
            # 이미 등장한 숫자라면
            counts[num] += 1 
            # 등장 횟수를 1 증가
        else:
            # 처음 등장한 숫자라면
            counts[num] = 1  
            # 등장 횟수를 1로 시작

    # 2단계: 가장 많이 등장한 숫자 찾기
    for num in counts:       
        # 등장한 숫자들만 하나씩 확인

        if counts[num] > max_count:
            # 현재 숫자의 등장 횟수가
            # 지금까지의 최대 횟수보다 크면

            max_count = counts[num]
            # 최대 등장 횟수 갱신

            mode = num
            # 가장 많이 등장한 숫자 갱신

    return mode
    # 최종적으로 가장 많이 등장한 숫자를 반환



############## 테스트 코드 삭제 금지 #################
print(find_mode([1, 2, 2, 3, 3, 3, 4]))  # 3
print(find_mode([5, 5, 6, 6, 5]))        # 5
#####################################################
```

```python
############## 주의 ##############
# input 함수는 사용하지 않습니다.
#
# [문제]
# orders 리스트에는 주문 상태가 순서대로 기록되어 있다.
# "PAID" 다음에 바로 "CANCEL"이 나온 횟수를 구하시오.
#
# [예시]
# ["PAID", "CANCEL", "PAID", "DONE"] → 1
#
# [요구조건]
# - 현재 상태와 다음 상태를 함께 비교한다
# 여기에 코드를 작성하여 함수를 완성합니다.
def count_paid_cancel(orders):
    result = 0

    for i in range(len(orders) - 1):            # 마지막 전까지만 순회
        if orders[i] == "PAID" and orders[i+1] == "CANCEL":
            result += 1                          # 조건 만족 시 카운트

    return result

# 내장 함수 제거
def count_paid_cancel(orders):
    result = 0
    prev = None                       # 이전 상태(현재 비교 기준)

    for current in orders:            # 리스트를 앞에서부터 순회
        if prev == "PAID" and current == "CANCEL":  # (이전,현재) 쌍이 PAID→CANCEL이면
            result += 1               # 횟수 1 증가
        prev = current                # 다음 비교를 위해 이전 상태 갱신

    return result


############## 테스트 코드 삭제 금지 #################
print(count_paid_cancel(["PAID", "CANCEL", "PAID", "DONE"]))      # 1
print(count_paid_cancel(["PAID", "PAID", "CANCEL", "CANCEL"]))   # 1
#####################################################
```

```python
############## 주의 ##############
# input 함수는 사용하지 않습니다.
#
# [문제]
# numbers 리스트에서
# 같은 값이 연속으로 등장한 최대 길이를 반환하시오.
#
# [예시]
# [1,1,2,2,2,3] → 3
#
# [요구조건]
# - 이전 값과 현재 값을 비교한다
# - 연속이 끊기면 길이를 초기화한다
def longest_same(numbers):
    # 첫 값 기준 세팅
    prev = numbers[0]
    count = 1
    max_count = 1

    # 두 번째 값부터 비교
    for num in numbers[1:]:
        if num == prev:
            count += 1
        else:
            count = 1  # 연속 끊김 → 초기화

        if count > max_count:
            max_count = count

        prev = num  # 이전 값 갱신

    return max_count

############## 테스트 코드 삭제 금지 #################
print(longest_same([1, 1, 2, 2, 2, 3]))  # 3
print(longest_same([5, 5, 5, 5]))        # 4
#####################################################
```

```python
############## 주의 ##############
# input 함수는 사용하지 않습니다.
#
# [문제]
# numbers 리스트에서 "이전 값보다 커진 경우"가 총 몇 번 발생했는지 반환하시오.
# 즉, i가 1부터 끝까지일 때,
# numbers[i] > numbers[i-1] 인 경우의 개수를 세어 반환합니다.
def count_increases(numbers):
    result = 0
    prev = numbers[0]           # 이전 값 저장

    for num in numbers[1:]:     # 두 번째 값부터 비교
        if num > prev:          # 현재 값이 이전 값보다 크면
            result += 1         # 증가 횟수 1 증가
        prev = num              # 이전 값 갱신

    return result


############## 테스트 코드 삭제 금지 #################
print(count_increases([1, 2, 3, 1, 2]))  # 3
print(count_increases([5, 4, 3, 2, 1]))  # 0
print(count_increases([1, 2, 3, 4, 5]))  # 4
#####################################################
```

```python
############## 주의 ##############
# input 함수는 사용하지 않습니다.
#
# [문제]
# numbers 리스트에서 "연속으로 증가하는 구간"의 최대 길이를 반환하시오.
# 단, "증가가 최소 1번 이상 발생한 구간"만 구간으로 인정합니다.
# 증가가 한 번도 없으면 0을 반환합니다.
#
# [요구조건]
# - 숫자가 이전 값보다 클 때만 증가로 판단 (num > prev)
# - 증가가 끊기면 현재 길이를 1로 초기화
# - 증가가 한 번도 없으면 0 반환
def longest_increasing_B(numbers):
    # 예외: 빈 리스트면 증가 구간이 있을 수 없으므로 0
    if not numbers:
        return 0

    max_len = 1         # (임시) 최대 길이 기록, 최소 1부터 시작
    current_len = 1     # 현재 연속 증가 구간 길이 (원소 개수)
    prev = numbers[0]   # 이전 값 저장
    has_increase = False  # 증가가 실제로 한 번이라도 있었는지 체크

    # 두 번째 원소부터 비교 시작
    for num in numbers[1:]:
        if num > prev:              # 증가라면
            current_len += 1        # 현재 구간 길이 증가
            has_increase = True     # 증가가 있었다고 표시
        else:                       # 증가가 끊겼다면(같거나 작아졌다면)
            current_len = 1         # 현재 구간 길이 초기화(새 구간 시작)

        # 최대 길이 갱신
        if current_len > max_len:
            max_len = current_len

        prev = num                  # 다음 비교를 위해 이전 값 갱신

    # 증가가 한 번도 없었다면(모든 구간이 길이 1뿐) 0 반환
    if not has_increase:
        return 0

    return max_len


############## 테스트 코드 삭제 금지 #################
print(longest_increasing_B([1, 2, 3, 1, 2]))  # 3
print(longest_increasing_B([5, 4, 3, 2, 1]))  # 0  (증가가 한 번도 없음)
print(longest_increasing_B([1, 2, 3, 4, 5]))  # 5
print(longest_increasing_B([2, 2, 2]))        # 0  (커진 적 없음)
#####################################################
```

```python
############## 주의 ##############
# input 함수는 사용하지 않습니다.
# 참고만 해라
# [문제]
# numbers 리스트에서
# "앞의 값 < 현재 값 > 다음 값" 패턴의 개수를 반환하시오.
#
# [예시]
# [1, 3, 2, 4, 1] → 2 (3, 4)
#
# [요구조건]
# - 현재 값 기준으로 양쪽 값을 비교한다
def count_peaks(numbers):
    # 1. 리스트 길이 직접 구하기
    length = 0
    for _ in numbers:
        length += 1

    count = 0
    i = 1  # 두 번째 요소부터 시작

    # 2. 앞 < 현재 > 뒤 비교
    while i < length - 1:
        if numbers[i - 1] < numbers[i] and numbers[i] > numbers[i + 1]:
            count += 1
        i += 1

    return count

############## 테스트 코드 삭제 금지 #################
print(count_peaks([1, 3, 2, 4, 1]))  # 2
print(count_peaks([1, 2, 3, 4]))     # 0
#####################################################
```

```python
############## 주의 ##############
# 복잡한 JSON 구조에서 데이터 수집/가공/출력 연습
# input 함수는 사용하지 않습니다.

data = {
    "meta": {"count": 4},
    "results": [
        {
            "id": 1,
            "user": {"name": "Kim", "email": "kim@test.com"},
            "company": {"name": "A-Corp", "location": {"city": "Seoul", "geo": {"lat": 37.56, "lng": 126.97}}},
            "tags": ["backend", "python"],
            "scores": [80, 90, 70],
        },
        {
            "id": 2,
            "user": {"name": "Lee", "email": "lee@test.com"},
            "company": {"name": "B-Lab", "location": {"city": "Busan", "geo": {"lat": 35.18, "lng": 129.07}}},
            "tags": ["frontend"],
            "scores": [60, 75],
        },
        {
            "id": 3,
            "user": {"name": "Park", "email": "park@test.com"},
            "company": {"name": "A-Corp", "location": {"city": "Seoul", "geo": {"lat": 37.50, "lng": 127.02}}},
            "tags": [],
            "scores": [100],
        },
        {
            "id": 4,
            "user": {"name": "Choi", "email": "choi@test.com"},
            "company": {"name": "C-Works", "location": {"city": "Incheon", "geo": {"lat": 37.45, "lng": 126.70}}},
            "tags": ["python", "data"],
            "scores": [88, 92, 95, 85],
        },
    ]
}

############## 주의 ##############
# 7) 출력 문자열 만들기 (가공 + 출력용 문자열 리스트)
# 각 사용자에 대해 아래 형식의 문자열을 만들어 리스트로 반환합니다.
# 형식: "{이름}({회사명}) - {도시} - tags:{태그개수}"
# input 함수는 사용하지 않습니다.
def make_summary_lines(data):
    lines = []

    for result in data["results"]:
        name = result["user"]["name"]
        company = result["company"]["name"]
        city = result["company"]["location"]["city"]

        # tags 개수 세기 (len 안 쓰고 카운트)
        tag_count = 0
        for _ in result["tags"]:
            tag_count += 1

        lines.append(f"{name}({company}) - {city} - tags:{tag_count}")

    return lines


############## 테스트 코드 삭제 금지 #################
print(make_summary_lines(data))
# [
#   'Kim(A-Corp) - Seoul - tags:2',
#   'Lee(B-Lab) - Busan - tags:1',
#   'Park(A-Corp) - Seoul - tags:0',
#   'Choi(C-Works) - Incheon - tags:2'
# ]
#####################################################
```
