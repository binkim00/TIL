```python
############## 주의 ##############
# 1부터 n까지의 합을 재귀로 구하기
# input 함수는 사용하지 않습니다.
# 반복문 사용 금지
def recursive_sum(n):
    if n == 1:
        return 1
    return n + recursive_sum(n - 1)

############## 테스트 코드 삭제 금지 #################
print(recursive_sum(5))  # 15
print(recursive_sum(1))  # 1
#####################################################
```

```python
############## 주의 ##############
# 팩토리얼 구하기 (재귀 사용)
# input 함수는 사용하지 않습니다.
# 반복문 사용 금지
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)

############## 테스트 코드 삭제 금지 #################
print(factorial(5))  # 120
print(factorial(1))  # 1
#####################################################
```

```python
############## 주의 ##############
# 리스트의 모든 요소 합 구하기 (재귀 사용)
# input 함수는 사용하지 않습니다.
# 반복문 사용 금지
def list_sum(numbers):
    if len(numbers) == 1:
        return numbers[0]
    return numbers[0] + list_sum(numbers[1:])

############## 테스트 코드 삭제 금지 #################
print(list_sum([1, 2, 3, 4]))  # 10
print(list_sum([5]))           # 5
#####################################################
```

```python
############## 주의 ##############
# 문자열 뒤집기 (재귀 사용)
# input 함수는 사용하지 않습니다.
# 반복문 사용 금지
def reverse_string(text):
    if len(text) == 1:
        return text
    return reverse_string(text[1:]) + text[0]

############## 테스트 코드 삭제 금지 #################
print(reverse_string("hello"))  # "olleh"
print(reverse_string("a"))      # "a"
#####################################################
```

```python
############## 주의 ##############
# 리스트에서 최댓값 찾기 (재귀 사용)
# input 함수는 사용하지 않습니다.
# 반복문 사용 금지
def find_max(numbers):
    if len(numbers) == 1:
        return numbers[0]
    
    sub_max = find_max(numbers[1:])
    if numbers[0] > sub_max:
        return numbers[0]
    else:
        return sub_max

############## 테스트 코드 삭제 금지 #################
print(find_max([3, 7, 2, 9, 4]))  # 9
print(find_max([1]))              # 1
#####################################################
```
